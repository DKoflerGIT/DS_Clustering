# kMeans++ Clusering Algorithm

#### Create random data points <br>
kMeans works for 2 and more dimensions. <br>
Plotting only works for 2 and 3 dimensions.

```python
from functions import kMeans as km
d = km.createRandomDatapoints(100, 2)
```

#### Find optimal number of clusters <br>
calculates the WCSS-score for various numbers of clusters and chooses the number, where the change in WCSS-score is less than 10%

```python
n = km.findOptimalClusterAmount(d)
```

#### Assign data points to a cluster<br>
Uses the kMeans++ algorithm to assign each datapoint to a cluster and optimizes all cluster-center-positions

```python
k = km.AssignClusters(d, n, True, 'k')
```

#### Plot result

```python
km.plotClusters(k, False)
print('WCSS-score: ' + str(km.withinClusterSumSquares(k)))
```
